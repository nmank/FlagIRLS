```python
n_its = 20
k=1
r=1
eps = .0000001
seed = 34

err = []
n = data[0].shape[0]


#initialize
np.random.seed(seed)
Y_raw = np.random.rand(n,r)-.5
Y = np.linalg.qr(Y_raw)[0][:,:r]
initialization = Y.copy()


#flag mean iteration function
itr = 1
diff = 1

als = []
while itr <= n_its and diff > 0.0000000001:
    Y0 = Y.copy()

    aX = []
    al = []

    ii=0

    for x in data:
        cossq = np.trace(Y0.T @ x @ x.T @ Y0)
        al.append((cossq+eps)**(-1/4))
        aX.append(al[-1]*x)
        ii+= 1
    als.append(al)

    aX = np.hstack(aX)

    Y = np.linalg.svd(aX)[0][:,:r]
    itr+=1

if diff > 0.0000000001:
    print('FlagIRLS not converged')
```

```python
weights = np.vstack(als).T
weights.shape
linestyles = ['dotted', 'dashed', 'dashdot', 'solid']

i=0
for w in weights:
    plt.plot(w, label = 'weight ' +str(i), linestyle = linestyles[i%4])
    i+=1
plt.xlabel('iteration')
plt.ylabel('weight value')
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
```
