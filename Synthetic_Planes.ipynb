```python
import numpy as np
import center_algorithms as ca
import matplotlib.pyplot as plt
```

```python
def calc_chordal_dist(X, Y, r):
    sum_sin = r- np.trace(X.T @ Y @ Y.T @ X)
    if sum_sin < 0:
        sum_sin = 0
    return np.sqrt(sum_sin)
```

```python
num_planes = 2
r=2
n_its = 10
seed = 1
```

```python
distances = []
for _ in range(1000):
    planes = []

    for _ in range(num_planes):
        planes.append(np.linalg.qr(np.random.rand(3,2))[0][:,:2])
    
    #calculate sine median
    sin_median, s_err = ca.irls_flag(planes, r, n_its, 'sine', opt_err = 'sine', fast = False, seed = seed)
    print('Sine Median finished')

    #calculate maximum cosine
    max_cosine, c_err = ca.irls_flag(planes, r, n_its, 'cosine', opt_err = 'cosine', fast = False, seed = seed)
    print('Max Cos finished')
    
    distances.append(calc_chordal_dist(max_cosine, sin_median, r))
```

```python
np.mean(distances)
```
